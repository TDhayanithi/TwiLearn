```python
grp = df.groupby('Category')
x = grp['Installs'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,'ro',color = 'orange')
plt.xticks(rotation = 90)
plt.title('Category vs Installs', fontsize = 25)
plt.xlabel('Categories',fontsize = 20)
plt.ylabel('Installs',fontsize = 20)
plt.show()
```

From this above graph,
1. Most peoples installs the  Category of 'Communication' app
2. Secondly most peoples installs the Category of 'Social' app

```python
plt.figure(figsize=(16,5))
plt.plot(y,'r--',color = 'g')
plt.xticks(rotation = 90)
plt.title('Category vs Price', fontsize = 25)
plt.xlabel('Categories',fontsize = 20)
plt.ylabel('Price',fontsize = 20)
plt.show()
```

From this above graph,
1. The Category of 'Finance' app has a higher price with a value of 2439 with rest of the categorical apps.
2. The second,third and fourth positions is taken by the category of 'Family','Lifestyle',and 'Medical'.

```python
plt.figure(figsize=(16,5))
plt.plot(z,'g^',color = 'r')
plt.xticks(rotation = 90)
plt.title('Category vs Reviews', fontsize = 25)
plt.xlabel('Categories',fontsize = 20)
plt.ylabel('Reviews',fontsize = 20)
plt.show()
```

From this above graph,

## Which are the top 10 expensive Apps in playstore?

```python
top10PaidApps = df[df['Type'] == 'Paid'].sort_values(by='Price', ascending=False).head(11)
# top10PaidApps
top10PaidApps_df = top10PaidApps[['App', 'Installs']]
plt.figure(figsize=(15,12));
plt.pie(top10PaidApps_df.Installs, explode=(0,0,0.25,0.5,0,0,0.5,0,0,0,0), labels=top10PaidApps_df.App, autopct='%1.1f%%', startangle=0);
plt.title('Top Expensive Apps Distribution',size = 20);
plt.legend(top10PaidApps_df.App, loc="lower right", title="Apps", fontsize = "xx-small");
```

From the above graph we can interprete that the App 'I am rich' is the most expensive app in the google playstore followed by 'I am Rich Premium'.

## What are the count of Apps in different genres?

```python
topAppsinGenres = df['Genres'].value_counts().head(50)
x3sis = []
y3sis = []

for i in range(len(topAppsinGenres)):
    x3sis.append(topAppsinGenres.index[i])
    y3sis.append(topAppsinGenres[i])
plt.figure(figsize=(15,9))
plt.ylabel('Genres(App Count)')
plt.xlabel('Genres')
graph = sns.barplot(x=x3sis,y=y3sis,palette="tab10")
graph.set_xticklabels(graph.get_xticklabels(), rotation=90, fontsize=12)
graph.set_title("Top Genres in the Playstore", fontsize = 20);
```

From the above visualization we can see that the Highest Number of Apps found in the Tools and Entertainment genres followed by Education, Action and many more.

```python
#we are going to look at the data with the stacked bars with pandas dataframe
#create count view per main_genre and content rating
df_by_main_genre = df.groupby(["Genres", "Content Rating"]).count().reset_index().sort_values(
    by=["App"], ascending=False).reset_index()

#select relevant columns
df_by_main_genre= df_by_main_genre[["Genres", "Content Rating", "App"]]

#reshape data to plot stacked bars
df_pivoted = df_by_main_genre.pivot(columns="Content Rating", index="Genres", values="App")
colors = ["#9b59b6","#2ecc71"]
df_pivoted.loc[:,df["Content Rating"].unique()].plot.bar(stacked=True, color=colors, figsize=(20,10))
plt.ylabel("Count", size = 20)
plt.xlabel('Generes', size = 20)
plt.title("Genres and Content Rating", size = 25)
```

Almost every app genre is for 'others'.
